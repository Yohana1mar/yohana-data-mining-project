```python
# range nya adalah 35 - 38
# diatas 36 dilarang masuk
# dibawah 35 suhu anda tidak normal silahkan masukkan kembali suhu
# diatas 38 suhu anda tidak normal silahkan masukkan kembali suhu

suhu = float(input("berapa suhu badan kamu? "))

if suhu >= 35 and suhu <= 36:
  print("silahkan masuk")
elif suhu < 35:
  print("tidak normal silahkan masukkan kembali suhu")
elif suhu > 38:
  print("tidak normal silahkan masukkan kembali suhu")
else:
  print('dilarang masuk')
```

seseorang yang ingin kursus data science offline, dimana harus melewati beberapa kondisi:

- ketika hujan deras maka kursus di cancel
- ketika hujan sedikit deras dan peserta ada minimal 5 maka kegiatan diadakan secara online
- ketika tidak hujan dan peserta ada minimal 5 maka kegiatan diadakan secara offline

```python
cuaca = input("cuara hari ini hujan deras, hujan sedikit deras atau tidak hujan?")
if cuaca == "hujan deras":
  print("kursus di cancel")
elif cuaca == "hujan sedikit deras":
  jumlah_peserta = int(input('jumlah beserta berapa? '))
  if jumlah_peserta >= 5:
    print("kursus diadakan secara online")
  else:
    print("kursus di cancel")
elif cuaca == "tidak hujan":
  jumlah_peserta = int(input('jumlah beserta berapa? '))
  if jumlah_peserta >= 5:
    print("kursus diadakan secara offline")
  else:
    print("kursus di cancel")
```

```python
cuaca = input("cuara hari ini hujan deras, hujan sedikit deras atau tidak hujan?")
if cuaca == "hujan deras":
  print("kursus di cancel")
elif cuaca == "hujan sedikit deras" or cuaca == "tidak hujan":
  jumlah_peserta = int(input('jumlah beserta berapa? '))
  if jumlah_peserta >= 5:
    if cuaca == "hujan sedikit deras":
      print("kursus diadakan secara online")
    elif cuaca == "tidak hujan":
      print("kursus diadakan secara offline")
  else:
    print("kursus di cancel")
```

## Latihan secara mandiri

Buat suatu logika ketika teman-teman berada di ATM untuk mengambil uang, buat variabel pin = 123456 dan total_saldo = bebas

```python
pin_benar = "123456"
total_saldo = 500000

input_pin = input("masukkan pin ATM anda ")

if input_pin == pin_benar:
    info_saldo = input("apakah anda ingin mengetahui informasi saldo? (iya/tidak) ")
    if info_saldo == "iya":
        print(f"Total saldo anda sekarang adalah {total_saldo}")

    tarik_tunai = input("apakah anda ingin tarik tunai? (iya/tidak) ")
    if tarik_tunai == "iya":
        nominal_tarik = int(input("berapa yang mau anda tarik? "))
        if nominal_tarik <= total_saldo:
            total_saldo -= nominal_tarik
            print(f"Selamat uang anda sudah ditarik senilai: {nominal_tarik} sisa saldo anda sekarang adalah {total_saldo}")
        else:
            print("Maaf, saldo anda tidak mencukupi.")
else:
    print("PIN yang anda masukkan salah.")
```
